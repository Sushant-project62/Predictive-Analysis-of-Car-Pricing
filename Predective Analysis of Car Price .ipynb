# **Data Cleaning**

# Checking the null values

```python
df.isnull().sum()
```

# Converting String into integer

```python
df['Engine Size (L)']=df['Engine Size (L)'].replace('Electric Motor',4)
```

```python
df['Engine Size (L)']=df['Engine Size (L)'].replace('1.5 + Electric',5)
```

# get the unique value of Engine Size Column

```python
df.describe()
```

# getting the mean median mode of the Engine size column

```python
df['Engine Size (L1)']=df['Engine Size (L)'].isna().sum()
```

```python
df['Engine Size (L1)'].unique()
```

```python
df['Engine Size (L1)'].describe()
```

# **Filling the null values with mode**

```python
mode_value = df['Engine Size (L)'].mode()
```

```python
mode_value
```

```python
df['Engine Size (L)'].fillna(mode_value, inplace=True)
```
